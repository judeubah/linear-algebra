We can use linear systems to solve matrices of linear equations.

So lets say we have the following matrix 

[a1x + b1y + c1z] = d1
[a2x + b2y + c2z] = d2
[a3x + b3y + c2z] = d3

we can see that this system also holds a series of sub-vectors that have been combined 

we have
    [a1]            [b1]         [c1]       [d1]
x*  [a2]    +   y * [b2]    + z* [c2]   =   [d2]
    [a3]            [b3]         [c3]       [d3]

these above vectors being multiplied by their corresponding scalars and then added to each other 


We want to get our matrices into the form of Ax = b
We can combine our vectors into a single matrix and then put our unknown values (x, y, z)
into a single vector

So long as our matrix has the row form of a, b, c and then the unknown values vector has the column form of [x] then when we multiply the matrix by this vector we will get 
               [y]  [a1x + b1y, + c1z ] 
               [z]  etc 
                    etc
This is because the unknown values are multipled on the right of the matrix and so the first column values of the matrix only hit the first row values of the unknown vector
so a always goes with x.
b always goes y
c always goes with z

and so now we have our linear system in the form Ax = b

so these two systems (below) are equivalent

[a1 b1 c1] [x]              [a1]        [b1]      [c1]  
[a2 b2 c2] [y]      === x * [a2]    +y* [b2] + z* [c2]  
[a3 b3 c3] [z]              [a3]        [b3]      [c3]  


After this we can get rid of the unknow variables and just inclued the input variables and the outputs

so
[a1, b1, c1| d1]
etc
etc
